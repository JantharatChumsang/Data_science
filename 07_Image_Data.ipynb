# 2. Plotting and Channels

```python
from skimage.io import imshow
import matplotlib.pyplot as plt
imshow(img) #โชว์ภาพ
plt.axis('off')
plt.show()
```

# 3. Point Operation

```python
import numpy as np
```

```python
from skimage.exposure import rescale_intensity
img1 = rescale_intensity(img, out_range=(100,255)).astype(np.uint8) #ปรับscale ละเอียดน้อยลงเหลือ100 จาก255 ภาพสว่างสูงสุดที่100
imshow(img1)
plt.axis('off')
plt.draw()
```

```python
from skimage.exposure import adjust_gamma
img1 = adjust_gamma(img, gamma=1.5).astype(np.uint8) #ปรับแกรมม่าเงา แสงสีคมชัด
imshow(img1)
plt.axis('off')
plt.draw()
```

```python
from skimage.exposure import adjust_gamma
img1 = adjust_gamma(img, gamma=0.5).astype(np.uint8)
imshow(img1)
plt.axis('off')
plt.draw()
```

# 4. Filtering

```python
from skimage.filters import gaussian
img1 = gaussian(img, sigma=3, multichannel=True) #ทำgaussian ให้ภาพเบลอปรับที่ sigma
imshow(img1)
plt.axis('off')
plt.draw()
```

```python
from skimage.filters import sobel, threshold_otsu  #ทำthreshold เป็นกรอบ/ขอบของภาพ
from skimage.color import rgb2gray
img0 = rgb2gray(img)
img1 = sobel(img0)
val = threshold_otsu(img1) # ถ้าไม่thresholdก็เห็นภาพแต่เป็นราง ๆ 
img2 = img1 > val
imshow(img2)
plt.axis('off')
plt.draw()
```
